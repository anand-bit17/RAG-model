```python
PINECONE_INDEX_NAME = "default"  # Replace with your desired index name

# Initialize Pinecone with the correct key and environment
pinecone.init(api_key=PINECONE_KEY, environment=PINECONE_ENV)

# List existing indexes
index_list = pinecone.list_indexes()

if len(index_list) == 0:
    # Create an index if none exists
    print("Creating index...")
    pinecone.create_index(PINECONE_INDEX_NAME, dimension=1536, metric='dotproduct')
else:
    print("Index already exists.")

# Print information about the index
index_info = pinecone.describe_index(PINECONE_INDEX_NAME)
print("Index information:", index_info)

# Create a Pinecone Index object
index = pinecone.Index(PINECONE_INDEX_NAME)
```
